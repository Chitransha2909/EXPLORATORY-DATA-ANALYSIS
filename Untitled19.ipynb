```python
plt.figure(figsize=(15,15))
g11=df.groupby(["Sub-Category"])[["Sales","Profit"]].mean()
g11.sort_values(by="Profit",ascending=False ).plot(kind="bar")
```

```python
g11.sum(axis=0)
```

```python
g11["sales%"] = g11["Sales"]/72.91
g11["Profit%"] =g11["Profit"]/10.95
g11.sort_values(by="Profit%",ascending=False)
g11.sort_values(by="sales%",ascending = False)
## COPIERS HAS HIGHEST PROFITS(74.69%) AND LOWEST PROFITS(-5%) TABLES
## COPIERS HAS HIGHEST SALES (30%) AND LOWEST SALES(0.1%) FASTENERS
```

#### Looking at the statistical figures, 
#### the company must have the main function in copiers since the sales and profits are heavily influenced by copiers.
#### The company must discontinue the sales of tables since it is contributing to a heavy loss and steps could be taken to improve profits for other sub categories
